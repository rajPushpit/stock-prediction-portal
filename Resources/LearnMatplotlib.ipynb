## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 38, 28] 
```

```python
plt.plot(days, original_price, 'b', label = 'Original Price')
plt.plot(days, predicted_price, 'r', label = 'Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

## Subplots

##### First subplot - original Price

```python
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label = 'Origianl Price')
plt.title('Original prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend();


# Second subplot - Predicted Price
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'r-o', label = 'Predicted Price')
plt.title('Predicted prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend();

plt.tight_layout()

# third subplot -  Difference between Original & pridicted Prices
```

```python
# third subplot -  Difference between Original & pridicted Prices
price_difference = [o-p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 1)
plt.bar(days, price_difference, color='purple', label = 'Price Difference')
plt.title('Difference between Original & Pridicted Prices')
plt.xlabel('Days')
plt.ylabel('Price Differnce')
plt.legend();
```

## Saving the Figure

```python
plt.plot(days, original_price, 'b', label = 'Original Price')
plt.plot(days, predicted_price, 'r', label = 'Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.savefig('examp_prediction.png', dpi =300)  # save the image
```
